```python
import pandas as pd
import numpy as np

df = pd.read_csv('Databaseforml/mumbai.csv')
df.head()
```

```python
#Dropping the redundant Columns
df = df.drop(['Rating Description','Tax'],axis = 'columns')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Dealing with Null Values

df['Rating'].fillna(0,inplace = True)
df['Reviews'].interpolate(method ='linear', limit_direction ='backward', inplace=True)
df.drop(['Star Rating'],axis = 'columns')
#Changing price variable as per our convenience
df['Price'] = df['Price'].str.replace(',','').astype(int)
d = {range(0, 2500):1, range(2500,5000):2, range(5000, 7500):3, range(7500, 12500):4,range(12500, 20000):5,range(25000,100000):6}
df['Price_point'] = df['Price']. apply(lambda x:next((v for k, v in d.items() if x in k),0))
```

```python
df.info()
```

```python
df.describe()
```

```python
import re

def convert_to_meters(value):
    if pd.isnull(value):
        return value
    
    if value == '0':
        return 0
    
    numeric_part = float(re.findall(r'[\d.]+', value)[0])
    unit = re.findall(r'[a-zA-Z]+', value)[0]
    
    if unit == 'km':
        return numeric_part * 1000
    elif unit == 'm':
        return numeric_part
    else:
        return None

df['Distance to Landmark'] = df['Distance to Landmark'].apply(convert_to_meters)

df.head()
```

```python
from sklearn.impute import SimpleImputer
imputer1 = SimpleImputer(strategy='mean')
#Rating-Mean
column_to_impute = df[['Rating']]
imputed_column = imputer1.fit_transform(column_to_impute)
df['Rating'] = imputed_column


imputer2=SimpleImputer(strategy='median')
column_to_impute = df[['Reviews']]
imputed_column = imputer2.fit_transform(column_to_impute)
df['Reviews'] = imputed_column

imputer3=SimpleImputer(strategy='median')
column_to_impute = df[['Distance to Landmark']]
imputed_column = imputer3.fit_transform(column_to_impute)
df['Distance to Landmark'] = imputed_column


df = df.drop(['Star Rating'],axis = 'columns')
```

```python
from sklearn.preprocessing import LabelEncoder

le1 = LabelEncoder()
df['Nearest Landmark']=le1.fit_transform(df['Nearest Landmark'])
df['Nearest Landmark'].fillna(0,inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder
le2 = LabelEncoder()
df['Location']= le2.fit_transform(df['Location'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(df.drop(['Price','Hotel Name','Price_point','Reviews'],axis = 'columns'),df['Price'],test_size = 0.3)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
X_train.head()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,f1_score
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier,XGBRegressor

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
mse_linear = mean_squared_error(y_test, y_pred_linear)
r2_linear = r2_score(y_test, y_pred_linear)
print("Linear Regression MSE:", mse_linear)
print("Linear Regression R-squared:", r2_linear)
```

```python
xgb_model = XGBRegressor()
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
mse_xgb = mean_squared_error(y_test, y_pred_xgb)
r2_xgb = r2_score(y_test, y_pred_xgb)
print("XGBoost MSE:", mse_xgb)
print("XGBoost R-squared:", r2_xgb)
```
